```python
import pandas as pd
import json

def read_jl(fn):
    with open(fn, 'r') as f:
        return pd.DataFrame([json.loads(d) for d in f.readlines()])
```

```python
df = read_jl('data/indeed-india/2018-07-12T09-41-38.jl')
```
